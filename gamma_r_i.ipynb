```python
import numpy as np

M_base = 1.0
f_base =0.5
alpha_base=0.5

M_high = 1.5 

M_low = 0.5 

r = 1.0 

Ratio_base = M_base*f_base/alpha_base
Ratio_high = M_high*f_base/alpha_base
Ratio_low = M_low*f_base/alpha_base

gamma = np.arange(11)
print(gamma)
len_gamma = len(gamma)

opt_ri_base= np.zeros(len_gamma)
print(opt_ri_base)
opt_ri_high= np.zeros(len_gamma)
#print(opt_ri_high)
opt_ri_low= np.zeros(len_gamma)
#print(opt_ri_low)
```

```python
from sympy import Eq, Symbol, solve

x = Symbol('x')

#### Solve for r_i with varying gamma under base ratio condition
for g in range(1,len_gamma):
    y=solve( ((1+g*x)**2) - ((g*Ratio_base)*(r-x))  )
    #print(g) 
    #print(y[1])
    opt_ri_base[g]= y[1]
    if (opt_ri_base[g]<0):
        opt_ri_base[g]= y[0]
```

```python
#### Solve for r_i with varying gamma under high ratio condition
for g in range(1,len_gamma):
    y=solve( ((1+g*x)**2) - ((g*Ratio_high)*(r-x))  )
    #print(g)
    #print(y[1])
    opt_ri_high[g]= y[1]
    if (opt_ri_high[g]<0):
        opt_ri_high[g]= y[0]
```

```python
#### Solve for r_i with varying gamma under low ratio condition
for g in range(1,len_gamma):
    y=solve( ((1+g*x)**2) - ((g*Ratio_low)*(r-x))  )
    #print(g)
    print('y[0]',y[0])
    print('y[1]',y[1])
    opt_ri_low[g]= y[1]
    if (opt_ri_low[g]<0):
        opt_ri_low[g]= y[0]  
    print(opt_ri_low[g])      
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(gamma,opt_ri_base,color='black',label='f=0.5,alpha=0.5, M=%0.2f'%M_base)
plt.scatter(gamma,opt_ri_high,color='magenta',label='f=0.5,alpha=0.5, M=%0.2f'%M_high)
plt.scatter(gamma,opt_ri_low,color='blue',label='f=0.5,alpha=0.5, M=%0.2f'%M_low)
plt.legend()
plt.xlabel('$\gamma$')
plt.ylabel('$r_i$*')
```

```python
from sympy import Eq, Symbol, solve

x = Symbol('x')

M = (np.arange(30)/10)
M = np.array([0.1,0.5,0.9,1.0,1.5,1.9,2.0,2.5])
M = np.array([1.0,1.5,4.0,8.5,10.00])
gamma= np.arange(11)
f=0.5
alpha=0.5

len_M=len(M)
len_gamma=len(gamma)

opt_ri= np.zeros((len_gamma,len_M ))

for g in range(1,len_gamma):
    for m in range(0,len_M):
        y=solve( ((1+gamma[g]*x)**2) - ((gamma[g]*M[m]*f/alpha)*(r-x))  )
        opt_ri[g][m]= y[1]
        
zero_line=np.zeros(len_gamma)        
```

```python
color=cm.rainbow(np.linspace(0,1,len_M))
plt.figure()
for m in range (0,len_M):
    plt.plot(gamma,opt_ri[:,m],c=color[m],label='$M$=%0.2f'%M[m])
    #plt.plot(gamma,zero_line,linewidth=5)
plt.legend(bbox_to_anchor=(1, 1.03))#, loc=1, borderaxespad=0)
#plt.legend(loc=4)
plt.title('f=%0.1f,alpha=%0.1f'%(f,alpha))
plt.xlabel('$\gamma$')
plt.ylabel('$r_i$*')
plt.ylim([-0.01,1.05])
save = '/Users/vandana/Dimorphism/manuscript/figures/ri_gamma_f=%0.1f,alpha=%0.1f.pdf'%(f,alpha)
#save = 'main_det_b_%s.pdf'%b
plt.savefig(save,format='pdf', bbox_inches='tight')
plt.show()    
```
